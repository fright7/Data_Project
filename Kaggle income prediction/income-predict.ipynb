```python
ca_loss_1 = ca_loss[ca_loss.capital_loss < 100]
```

```python
ca_loss_list = ca_loss_1.index.to_list()

ca_loss_index = []

for i in range(len(df7.capital_loss)):
    
    if df7.capital_loss[i] in ca_loss_list:
        
        ca_loss_index.append(i)
        
df8 = df7.drop(index=ca_loss_index,axis=1)
```

```python
df_str = df9.select_dtypes(exclude=["number"])
str_val = df_str.columns.to_list()
```

```python
df_str
```

```python
label_encode = LabelEncoder()
```

```python
df_str2 = df_str.apply(label_encode.fit_transform)
```

```python
df_str2
```

```python
df9_number = df9.drop(labels=str_val,axis=1)
```

```python
df10 = pd.concat([df9_number,df_str2],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max = MinMaxScaler()


min_max_list = ["capital_gain","capital_loss"]

min_max_df = df10.loc[:,min_max_list]

fnlwgt_log = np.log(df10.fnlwgt)

min_max_df2 = min_max_df.join(fnlwgt_log)

min_max_df3 = pd.DataFrame(min_max.fit_transform(min_max_df2),columns=min_max_df2.columns)



df10_drop = df10.drop(labels=["fnlwgt","capital_gain","capital_loss"],axis=1)

df10_drop.reset_index(drop=True,inplace=True)
```

```python
df11 = pd.concat([min_max_df3,df10_drop],axis=1)
```

```python
train_df = df11.drop(labels=["id"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
```

```python
X = train_df.iloc[:,:-1]
y = train_df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=62)
```

```python
model1 = RandomForestClassifier()

model1.fit(X_train,y_train)
model1_pred = model1.predict(X_test)
accuracy1 = accuracy_score(y_test,model1_pred)
f1_1 = f1_score(y_test,model1_pred)
print(accuracy1,f1_1)
```

```python
model2 = XGBClassifier(tree_method = "hist")

model2.fit(X_train,y_train)
model2_pred = model2.predict(X_test)
accuracy2 = accuracy_score(y_test,model2_pred)
f1_2 = f1_score(y_test,model2_pred)
print(accuracy2,f1_2)
```

```python
scorer = make_scorer(accuracy_score)


param = {

    "max_depth" : [30,50,70],
    "eta" : [0.05,0.1]

}

gs = GridSearchCV(estimator = model2,param_grid = param,scoring=scorer,n_jobs=4,verbose=10)
gs.fit(X_train.values,y_train.values)
```

```python
model_gs = gs.best_estimator_
gs_pred = model_gs.predict(X_test.values)
gs_score = accuracy_score(y_test,gs_pred)
print(gs_score)
```
